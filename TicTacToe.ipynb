```python
#zadanie domowe 

print("Witaj w grze KOLKO I KRZYZYK")

board = ["-", "-", "-","-", "-", "-","-", "-", "-"]
    

gracz = "X"
wygrany = None
gra_trwa = True

def print_board():
    print("".join(board[:3]))
    print("".join(board[3:6]))
    print("".join(board[6:9]))  
    
    
def zagraj():
    print_board()
    
    while gra_trwa:
        
        ruch(gracz)
        
        sprawdz_koniec()
        
        zmien_gracza()
        
#koniec gry
    if wygrany == "X" or wygrany =="O":
        print("Gratulacje! " + wygrany + " wygral!")
    elif wygrany == None:
        print("Remis.")
    
def ruch(gracz):
    
    print("Kolej na na gracza: " + gracz)
    gdzie_x = input("Wybierz pole od 1 do 9: ")
    
    
    valid = False 
    while not valid:
        while gdzie_x not in ["1", "2", "3", "4" ,"5", "6", "7", "8", "9"]:
            gdzie_x = input("Blad! Wybierz pole od 1 do 9: ")
            
        gdzie_x = int(gdzie_x) - 1

        if board[gdzie_x] == "-":
            valid = True
        else:
            print("Pole zajete! Sprobuj jeszcze raz.")
    
    board[gdzie_x] = gracz
    print_board()
    
def sprawdz_koniec():
    wygrana()
    remis()
    
    
def wygrana():
    global wygrany
    #sprawdzenie wierszy kolumn i na skos
    wygrana_wiersze = wiersze()
    wygrana_kolumny = kolumny()
    wygrana_naskos = naskos()
    
    if wygrana_wiersze:
        wygrany = wygrana_wiersze
    elif wygrana_kolumny:
        wygrany = wygrana_kolumny
    elif wygrana_naskos:
        wygrany = wygrana_naskos
    else:
        #nie ma wygranej
        wygrany = None
    return

def wiersze():
    global gra_trwa
    wiersz_1 = board[0] == board[1] == board[2] != "-"
    wiersz_2 = board[3] == board[4] == board[5] != "-"
    wiersz_3 = board[6] == board[7] == board[8] != "-"
    
    if wiersz_1 or wiersz_2 or wiersz_3:
        gra_trwa = False
        
    #pokaz zwyciezce
    if wiersz_1:
        return board[0]
    elif wiersz_2:
        return board[3]
    elif wiersz_3:
        return board[6]
    return

def kolumny():
    global gra_trwa
    kolumna_1 = board[0] == board[3] == board[6] != "-"
    kolumna_2 = board[1] == board[4] == board[7] != "-"
    kolumna_3 = board[2] == board[5] == board[8] != "-"
    
    if kolumna_1 or kolumna_2 or kolumna_3:
        gra_trwa = False
        
    #pokaz zwyciezce
    if kolumna_1:
        return board[0]
    elif kolumna_2:
        return board[1]
    elif kolumna_3:
        return board[2]
    return

def naskos():
    global gra_trwa
    naskos_1 = board[2] == board[4] == board[6] != "-"
    naskos_2 = board[0] == board[4] == board[8] != "-"
  
    if naskos_1 or naskos_2:
        gra_trwa = False
        
    #pokaz zwyciezce
    if naskos_1:
        return board[0]
    elif naskos_2:
        return board[2]

    return
    
def remis():
    global gra_trwa
    if "-" not in board:
        gra_trwa = False
    return

def zmien_gracza():
    global gracz
    
    if gracz == "X":
        gracz = "O"
    elif gracz == "O":
        gracz = "X"
    return

zagraj()
```
